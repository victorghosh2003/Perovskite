```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV 
```

```python
BC = df = pd.read_csv('JMXLWMIFDJCGBV-UHFFFAOYSA-N - Sheet1classifier.csv')
pd.set_option('display.max_rows',800) # lines 1,2,3 make sure that full output is shown
pd.set_option('display.max_columns',800) # lines 1,2,3 make sure that full output is shown
pd.set_option('display.width',1000)
print(BC)
```

```python
X = df.drop(['crystal_score','rxn_m_organic','rxn_m_acid','rxn_m_inorganic','ln_K','classifier'], axis='columns')
y = df.classifier
print(X)
```

```python
for x in range (150):    
    clf = svm.SVC()
    clf.fit(X, y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    model = SVC(kernel='rbf')
    model.fit(X_train, y_train)
    data = (model.score(X_test, y_test))
    #print(data)
(data.mean()) 


```

```python
from sklearn.model_selection import GridSearchCV 
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001, 10, 100, 1000], 
              'kernel': ['rbf', 'linear']}  
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3) 
  
# fitting the model for grid search 
grid.fit(X_train, y_train) 
```

```python
print(grid.best_params_) 
print(grid.best_estimator_) 
```

```python
for x in range (1):
    for x in range (300):    
        clf = svm.SVC()
        clf.fit(X, y)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
        model = SVC(C=100, break_ties=False, cache_size=200, class_weight=None, coef0=0.0,
        decision_function_shape='ovr', degree=3, gamma=0.1, kernel='rbf',
        max_iter=-1, probability=False, random_state=None, shrinking=True,
        tol=0.001, verbose=False)
        model.fit(X_train, y_train)
        data = (model.score(X_test, y_test))
        #print(data)
    m = (data.mean()) #stdev is 0.03536101821 python stdev function gives me 0 for some reason, calculated using excel.
    print(m)



```

```python
y_pred = clf.predict(X)
print(y_pred)
```
